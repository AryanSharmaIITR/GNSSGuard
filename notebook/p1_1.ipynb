```python
import pandas as pd
import os
```

```python
def split_timeseries_csv(
    name,
    train_path,
    val_path,
    date_column=None,   # optional
    train_ratio=0.7
):
    # ---- Load data ----
    df = pd.read_csv(f"{train_path}/train_{name}.csv")

    # ---- Split index ----
    n = len(df)
    split_idx = int(n * train_ratio)

    train_df = df.iloc[:split_idx]
    val_df = df.iloc[split_idx:]

    # ---- Ensure directory exists ----
    os.makedirs(os.path.dirname(train_path), exist_ok=True)
    os.makedirs(os.path.dirname(val_path), exist_ok=True)

    # ---- Save ----
    train_df.to_csv(f"{train_path}/train_{name}.csv", index=False)
    val_df.to_csv(f"{val_path}/val_{name}.csv", index=False)

    print("✅ Split completed")
    print(f"Total rows : {n}")
    print(f"Train rows : {len(train_df)}")
    print(f"Val rows   : {len(val_df)}")
    print(f"Train saved to: {train_path}")
    print(f"Val saved to  : {val_path}")

```

```python
train_base="../dataset/train"
val_base="../dataset/val"
```

```python
for i in range(0,8):
    split_timeseries_csv(f"ch{i}",train_base,val_base)
```
