```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import stft, spectrogram, welch
from scipy import stats
from scipy.ndimage import gaussian_filter1d
import warnings
warnings.filterwarnings('ignore')

class GNSSSignalAnalyzer:
    """
    Enhanced GNSS signal analysis specifically designed for available observables:
    - PIP/PQP: In-phase/Quadrature Prompt correlator outputs (ideal for spoofing detection)
    - CN0: Carrier-to-noise ratio
    - Carrier_phase: Phase measurements for cycle slip detection
    - TCD: Tracking channel status
    """
    
    def __init__(self, df):
        """
        Initialize analyzer with GNSS data.
        
        Parameters:
        -----------
        df : pandas.DataFrame
            GNSS observables with columns: PRN, time, Carrier_Doppler_hz, 
            Pseudorange_m, PIP, PQP, CN0, Carrier_phase, TCD, spoofed
        """
        self.df = df.copy()
        self._validate_data()
        self.l1_freq = 1575.42e6  # GPS L1 frequency in Hz
        self.c = 299792458.0  # Speed of light in m/s
        
    def _validate_data(self):
        """Ensure required columns exist."""
        required = ['PRN', 'time', 'Carrier_Doppler_hz', 'Pseudorange_m']
        missing = [col for col in required if col not in self.df.columns]
        if missing:
            raise ValueError(f"Missing required columns: {missing}")
    
    def construct_iq_signal(self, signal_df):
        """
        Construct complex I/Q signal from PIP (In-phase) and PQP (Quadrature) correlator outputs.
        This is the most authentic representation of the received GNSS signal.
        
        Returns:
        --------
        complex_signal : array of complex values
        iq_power : total correlator power
        phase_angle : phase angle between I and Q branches
        """
        if 'PIP' in signal_df.columns and 'PQP' in signal_df.columns:
            # Clean and normalize correlator outputs
            pip = signal_df['PIP'].astype(float).values
            pqp = signal_df['PQP'].astype(float).values
            
            # Remove outliers
            pip = self._remove_outliers(pip)
            pqp = self._remove_outliers(pqp)
            
            # Construct complex baseband signal
            complex_signal = pip + 1j * pqp
            
            # Compute derived metrics
            iq_power = np.abs(complex_signal) ** 2
            phase_angle = np.angle(complex_signal)
            
            return complex_signal, iq_power, phase_angle
        else:
            # Fallback to Doppler + Pseudorange construction
            doppler = signal_df['Carrier_Doppler_hz'].values
            pr = signal_df['Pseudorange_m'].values
            doppler_norm = (doppler - np.median(doppler)) / (np.std(doppler) + 1e-6)
            pr_norm = (pr - np.median(pr)) / (np.std(pr) + 1e-6)
            return doppler_norm + 1j * pr_norm, None, None
    
    def _remove_outliers(self, data, threshold=3):
        """Remove outliers using median absolute deviation."""
        median = np.median(data)
        mad = np.median(np.abs(data - median))
        modified_z_scores = 0.6745 * (data - median) / (mad + 1e-6)
        data_clean = data.copy()
        data_clean[np.abs(modified_z_scores) > threshold] = median
        return data_clean
    
    def compute_enhanced_features(self, signal_segment):
        """
        Compute domain-specific features for GNSS signal characterization.
        Leverages PIP, PQP, CN0, and Carrier_phase for robust analysis.
        """
        features = {}
        
        # 1. I/Q Correlator Analysis (most powerful for spoofing detection)
        if 'PIP' in signal_segment.columns and 'PQP' in signal_segment.columns:
            pip = signal_segment['PIP'].values
            pqp = signal_segment['PQP'].values
            
            # Correlator power stability
            power = pip**2 + pqp**2
            features['iq_power_mean'] = np.mean(power)
            features['iq_power_std'] = np.std(power)
            features['iq_power_cv'] = features['iq_power_std'] / (features['iq_power_mean'] + 1e-6)
            
            # Phase distribution (authentic: uniform, spoofed: often clustered)
            phase = np.arctan2(pqp, pip)
            features['phase_circular_mean'] = stats.circmean(phase)
            features['phase_circular_std'] = stats.circstd(phase)
            
            # I/Q correlation (should be low for authentic signals)
            features['iq_correlation'] = np.corrcoef(pip, pqp)[0, 1]
            
            # Symmetry metric (authentic: symmetric scatter)
            features['iq_skewness'] = (stats.skew(pip) + stats.skew(pqp)) / 2
            features['iq_kurtosis'] = (stats.kurtosis(pip) + stats.kurtosis(pqp)) / 2
        
        # 2. CN0 Analysis
        if 'CN0' in signal_segment.columns:
            cn0 = signal_segment['CN0'].values
            features['cn0_mean'] = np.mean(cn0)
            features['cn0_std'] = np.std(cn0)
            features['cn0_min'] = np.min(cn0)
            features['cn0_max'] = np.max(cn0)
            features['cn0_drop_rate'] = (cn0[0] - cn0[-1]) / len(cn0) if len(cn0) > 1 else 0
            
            # CN0 stability (spoofed signals often have unusually stable CN0)
            if len(cn0) > 10:
                features['cn0_stability'] = 1.0 / (np.std(cn0) + 1e-6)
            else:
                features['cn0_stability'] = 0
        
        # 3. Doppler and Pseudorange Analysis
        doppler = signal_segment['Carrier_Doppler_hz'].values
        pr = signal_segment['Pseudorange_m'].values
        times = signal_segment['time'].values
        
        if len(doppler) > 2:
            # Doppler rate (should be smooth for authentic signals)
            doppler_rate = np.diff(doppler) / np.diff(times)
            features['doppler_rate_mean'] = np.mean(doppler_rate)
            features['doppler_rate_std'] = np.std(doppler_rate)
            features['doppler_rate_rms'] = np.sqrt(np.mean(doppler_rate**2))
            
            # Pseudorange rate consistency with Doppler
            pr_rate = np.diff(pr) / np.diff(times)
            expected_doppler = -pr_rate * self.l1_freq / self.c
            if len(expected_doppler) > 0:
                features['pr_doppler_consistency'] = np.corrcoef(
                    doppler[1:], expected_doppler
                )[0, 1]
            
            # Doppler smoothness (spoofed signals may show unnatural patterns)
            features['doppler_jerk'] = np.mean(np.abs(np.diff(doppler_rate)))
            features['doppler_linearity'] = self._check_linearity(doppler)
        
        # 4. Carrier Phase Analysis (cycle slip detection)
        if 'Carrier_phase' in signal_segment.columns:
            phase = signal_segment['Carrier_phase'].values
            if len(phase) > 3:
                # Phase acceleration (indicator of dynamics)
                phase_diff2 = np.diff(phase, n=2)
                features['phase_acceleration_mean'] = np.mean(np.abs(phase_diff2))
                features['phase_acceleration_max'] = np.max(np.abs(phase_diff2))
                
                # Cycle slip detection (sudden phase jumps)
                phase_diff = np.diff(phase)
                phase_diff_normalized = (phase_diff - np.median(phase_diff)) / (np.std(phase_diff) + 1e-6)
                features['cycle_slip_count'] = np.sum(np.abs(phase_diff_normalized) > 5)
                features['cycle_slip_intensity'] = np.max(np.abs(phase_diff_normalized))
        
        # 5. TCD (Tracking Channel Status) Analysis
        if 'TCD' in signal_segment.columns:
            tcd = signal_segment['TCD'].values
            features['tracking_stability'] = 1.0 - np.std(tcd) / (np.mean(tcd) + 1e-6)
            features['tcd_changes'] = np.sum(np.diff(tcd) != 0)
        
        # 6. Overall Spoofing Likelihood Score
        features['spoofing_likelihood'] = self._compute_spoofing_score(features)
        
        return features
    
    def _check_linearity(self, signal):
        """Check if signal follows linear trend (authentic signal characteristic)."""
        t = np.arange(len(signal))
        linear_fit = np.polyfit(t, signal, 1)
        residual = signal - np.polyval(linear_fit, t)
        r2 = 1 - np.var(residual) / (np.var(signal) + 1e-6)
        return max(0, min(1, r2))
    
    def _compute_spoofing_score(self, features):
        """
        Compute spoofing likelihood score based on multiple indicators.
        Returns value between 0 (authentic) and 1 (spoofed).
        """
        score = 0
        weight_sum = 0
        
        # CN0 unusually stable (spoofing indicator)
        if 'cn0_stability' in features:
            if features['cn0_stability'] > 10:  # Threshold based on typical values
                score += min(1, features['cn0_stability'] / 20)
                weight_sum += 1
        
        # I/Q correlation too high (spoofing indicator)
        if 'iq_correlation' in features:
            if abs(features['iq_correlation']) > 0.3:
                score += min(1, abs(features['iq_correlation']))
                weight_sum += 1
        
        # I/Q power unusually stable
        if 'iq_power_cv' in features:
            if features['iq_power_cv'] < 0.1:  # Coefficient of variation < 10%
                score += 1 - features['iq_power_cv'] / 0.1
                weight_sum += 1
        
        # Phase not uniform (spoofing indicator)
        if 'phase_circular_std' in features:
            # Authentic should have std ~ 1.8 rad (uniform distribution)
            phase_std_deviation = abs(features['phase_circular_std'] - 1.8)
            if phase_std_deviation > 0.5:
                score += min(1, phase_std_deviation / 1.8)
                weight_sum += 1
        
        # Doppler too smooth/unnatural
        if 'doppler_linearity' in features:
            if features['doppler_linearity'] > 0.99:  # Unusually linear
                score += features['doppler_linearity'] * 0.5
                weight_sum += 0.5
        
        # PR-Doppler inconsistency
        if 'pr_doppler_consistency' in features:
            if features['pr_doppler_consistency'] < 0.7:
                score += 1 - features['pr_doppler_consistency']
                weight_sum += 1
        
        return score / weight_sum if weight_sum > 0 else 0.5
    
    def graph_enhanced_stft(self, starting_index=0, ending_index=128, spoofed=None, 
                           signal_type='iq', use_cn0_weighting=True):
        """
        Enhanced STFT visualization with GNSS-specific processing.
        
        Parameters:
        -----------
        starting_index : int
            Starting index for data selection
        ending_index : int
            Ending index for data selection (exclusive)
        spoofed : bool or None
            Filter by spoofing status (if column exists)
        signal_type : str
            'iq' for PIP/PQP based, 'doppler', 'pseudorange', 'complex'
        use_cn0_weighting : bool
            Weight signal by CN0 for better quality assessment
        """
        # Filter data by indices
        signal_df = self.df.sort_values('time').iloc[starting_index:ending_index].copy()
        
        # Apply spoofed filter if specified and column exists
        if spoofed is not None and 'spoofed' in signal_df.columns:
            signal_df = signal_df[signal_df['spoofed'] == spoofed]
            
        if signal_df.empty:
            print(f"No data found for indices {starting_index}:{ending_index}")
            return None
        
        # Get PRN for display (if available)
        prn_display = ""
        if 'PRN' in signal_df.columns and len(signal_df['PRN'].unique()) == 1:
            prn_display = f"PRN {signal_df['PRN'].iloc[0]} | "
        
        # Compute enhanced features
        features = self.compute_enhanced_features(signal_df)
        
        # Extract signals
        times = signal_df['time'].astype(float).values
        dt = np.median(np.diff(times))
        fs = 1.0 / dt if dt > 0 else 10.0
        
        # Construct signal based on type
        if signal_type == 'iq' and 'PIP' in signal_df.columns:
            signal, iq_power, phase = self.construct_iq_signal(signal_df)
            ylabel = 'I/Q Magnitude'
            signal_for_stft = signal
        elif signal_type == 'doppler':
            signal_for_stft = signal_df['Carrier_Doppler_hz'].values
            ylabel = 'Doppler Frequency (Hz)'
        elif signal_type == 'pseudorange':
            signal_for_stft = signal_df['Pseudorange_m'].values
            ylabel = 'Pseudorange (m)'
        else:
            # Complex combination
            doppler = signal_df['Carrier_Doppler_hz'].values
            pr = signal_df['Pseudorange_m'].values
            doppler_norm = (doppler - np.median(doppler)) / (np.std(doppler) + 1e-6)
            pr_norm = (pr - np.median(pr)) / (np.std(pr) + 1e-6)
            signal_for_stft = doppler_norm + 1j * pr_norm
            ylabel = 'Normalized Complex Signal'
        
        # Apply CN0 weighting if available
        if use_cn0_weighting and 'CN0' in signal_df.columns:
            weights = signal_df['CN0'].values / signal_df['CN0'].max()
            if len(signal_for_stft.shape) > 0:
                signal_for_stft = signal_for_stft * weights
        
        # Adaptive STFT parameters
        nperseg = min(max(16, len(signal_for_stft) // 4), len(signal_for_stft))
        noverlap = nperseg * 3 // 4
        
        # Compute STFT
        f, t, Zxx = stft(signal_for_stft, fs=fs, nperseg=nperseg, 
                        noverlap=noverlap, window='hamming',
                        boundary='even', padded=True)
        
        # Create comprehensive visualization
        fig = plt.figure(figsize=(16, 10))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
        
        title_text = f'GNSS Signal Analysis - {prn_display}'
        if spoofed is not None:
            title_text += f'{"⚠️ SPOOFED" if spoofed else "✓ AUTHENTIC"}'
        else:
            # Check actual spoofed status if column exists
            if 'spoofed' in signal_df.columns and len(signal_df) > 0:
                actual_spoofed = signal_df['spoofed'].iloc[0]
                title_text += f'{"⚠️ SPOOFED" if actual_spoofed else "✓ AUTHENTIC"}'
        
        title_text += f'\nSpoofing Likelihood: {features.get("spoofing_likelihood", 0):.3f}'
        fig.suptitle(title_text, fontsize=14, fontweight='bold')
        
        # 1. STFT Spectrogram (Top-left, spanning 2 columns)
        ax1 = fig.add_subplot(gs[0, :2])
        Zxx_db = 20 * np.log10(np.abs(Zxx) + 1e-10)
        im1 = ax1.pcolormesh(t, f, Zxx_db, shading='gouraud', cmap='viridis')
        ax1.set_ylabel('Frequency (Hz)')
        ax1.set_xlabel('Time (s)')
        ax1.set_title('STFT Spectrogram (dB)')
        plt.colorbar(im1, ax=ax1, label='Magnitude (dB)')
        ax1.grid(True, alpha=0.3)
        
        # 2. I/Q Scatter Plot (Top-right)
        ax2 = fig.add_subplot(gs[0, 2])
        if 'PIP' in signal_df.columns and 'PQP' in signal_df.columns:
            pip = signal_df['PIP'].values
            pqp = signal_df['PQP'].values
            scatter = ax2.scatter(pip, pqp, c=times, cmap='plasma', alpha=0.6, s=30)
            ax2.axhline(y=0, color='k', linestyle='--', alpha=0.3)
            ax2.axvline(x=0, color='k', linestyle='--', alpha=0.3)
            ax2.set_xlabel('In-Phase (PIP)')
            ax2.set_ylabel('Quadrature (PQP)')
            ax2.set_title('I/Q Correlator Constellation')
            ax2.set_aspect('equal')
            plt.colorbar(scatter, ax=ax2, label='Time (s)')
            ax2.grid(True, alpha=0.3)
        else:
            ax2.text(0.5, 0.5, 'I/Q Data Not Available', 
                    ha='center', va='center', transform=ax2.transAxes)
            ax2.set_xlim(0, 1)
            ax2.set_ylim(0, 1)
        
        # 3. Time Domain Signals (Middle row, spanning 2 columns)
        ax3 = fig.add_subplot(gs[1, :2])
        ax3_twin = ax3.twinx()
        
        doppler = signal_df['Carrier_Doppler_hz'].values
        pr = signal_df['Pseudorange_m'].values
        
        line1 = ax3.plot(times, doppler, 'b-', label='Doppler', linewidth=1.5, alpha=0.8)
        line2 = ax3_twin.plot(times, pr, 'r--', label='Pseudorange', linewidth=1.5, alpha=0.8)
        
        ax3.set_xlabel('Time (s)')
        ax3.set_ylabel('Doppler (Hz)', color='b')
        ax3_twin.set_ylabel('Pseudorange (m)', color='r')
        ax3.set_title('Time Domain Observables')
        ax3.grid(True, alpha=0.3)
        
        lines = line1 + line2
        labels = [l.get_label() for l in lines]
        ax3.legend(lines, labels, loc='upper left')
        
        # 4. CN0 Plot (Middle-right)
        ax4 = fig.add_subplot(gs[1, 2])
        if 'CN0' in signal_df.columns:
            cn0 = signal_df['CN0'].values
            ax4.plot(times, cn0, 'g-', linewidth=2)
            ax4.fill_between(times, cn0, cn0.min(), alpha=0.3)
            ax4.set_xlabel('Time (s)')
            ax4.set_ylabel('CN0 (dB-Hz)')
            ax4.set_title('Carrier-to-Noise Ratio')
            ax4.grid(True, alpha=0.3)
            ax4.axhline(y=cn0.mean(), color='r', linestyle='--', 
                       label=f'Mean: {cn0.mean():.1f}')
            ax4.legend()
        
        # 5. Feature Comparison Radar (Bottom-left)
        ax5 = fig.add_subplot(gs[2, 0], projection='polar')
        # Determine spoofed status for radar color
        radar_spoofed = spoofed
        if radar_spoofed is None and 'spoofed' in signal_df.columns and len(signal_df) > 0:
            radar_spoofed = signal_df['spoofed'].iloc[0]
        self._plot_feature_radar(ax5, features, radar_spoofed)
        
        # 6. Phase Analysis (Bottom-middle)
        ax6 = fig.add_subplot(gs[2, 1])
        if 'Carrier_phase' in signal_df.columns:
            phase = signal_df['Carrier_phase'].values
            phase_unwrapped = np.unwrap(phase)
            ax6.plot(times, phase_unwrapped, 'm-', linewidth=1.5)
            ax6.set_xlabel('Time (s)')
            ax6.set_ylabel('Carrier Phase (cycles)')
            ax6.set_title('Carrier Phase Evolution')
            ax6.grid(True, alpha=0.3)
        else:
            ax6.text(0.5, 0.5, 'Phase Data Not Available', 
                    ha='center', va='center', transform=ax6.transAxes)
        
        # 7. Statistics Summary (Bottom-right)
        ax7 = fig.add_subplot(gs[2, 2])
        ax7.axis('off')
        stats_text = self._format_feature_stats(features, radar_spoofed)
        ax7.text(0.05, 0.95, stats_text, transform=ax7.transAxes,
                fontsize=9, verticalalignment='top', fontfamily='monospace',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        
        plt.tight_layout()
        plt.show()
        
        # Print analysis summary
        print(f"\n{'='*50}")
        print(f"GNSS SIGNAL ANALYSIS SUMMARY - {prn_display}")
        print(f"{'='*50}")
        print(f"Data Range: indices {starting_index} to {ending_index} ({len(signal_df)} samples)")
        print(f"Sampling: {fs:.2f} Hz | Window: {nperseg} samples ({nperseg/fs:.3f}s)")
        print(f"STFT Resolution: Δt={t[1]-t[0]:.3f}s, Δf={f[1]-f[0]:.3f}Hz")
        print(f"Spoofing Likelihood Score: {features.get('spoofing_likelihood', 0):.3f}")
        
        if features.get('spoofing_likelihood', 0) > 0.7:
            print("⚠️  HIGH PROBABILITY OF SPOOFING DETECTED")
        elif features.get('spoofing_likelihood', 0) > 0.4:
            print("⚡ MODERATE SPOOFING INDICATORS")
        else:
            print("✓ SIGNAL APPEARS AUTHENTIC")
        print(f"{'='*50}\n")
        
        return {'f': f, 't': t, 'Zxx': Zxx, 'features': features}
    
    def _plot_feature_radar(self, ax, features, spoofed):
        """Create radar chart for feature visualization."""
        # Select key features for radar (normalized indicators)
        radar_features = {
            'CN0 Stability': features.get('cn0_stability', 0) / 20,
            'I/Q Correlation': 1 - min(1, abs(features.get('iq_correlation', 0))),
            'Doppler Smoothness': features.get('doppler_linearity', 0),
            'PR-Doppler Consist': max(0, features.get('pr_doppler_consistency', 0)),
            'Phase Continuity': 1 - min(1, features.get('cycle_slip_intensity', 0) / 10),
            'Tracking Quality': features.get('tracking_stability', 0.5)
        }
        
        categories = list(radar_features.keys())
        values = list(radar_features.values())
        
        angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
        values += values[:1]
        angles += angles[:1]
        
        ax.set_theta_offset(np.pi / 2)
        ax.set_theta_direction(-1)
        ax.plot(angles, values, 'o-', linewidth=2, 
                color='green' if spoofed is False else 'red' if spoofed is True else 'blue')
        ax.fill(angles, values, alpha=0.25, 
                color='green' if spoofed is False else 'red' if spoofed is True else 'blue')
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(categories, fontsize=8)
        ax.set_ylim(0, 1)
        ax.set_title('Signal Quality Metrics', pad=20)
        ax.grid(True)
    
    def _format_feature_stats(self, features, spoofed):
        """Format features for display."""
        if spoofed is True:
            status = "SPOOFED"
        elif spoofed is False:
            status = "AUTHENTIC"
        else:
            status = "UNKNOWN"
            
        text = f"{status} SIGNAL STATS\n"
        text += "="*28 + "\n"
        
        key_metrics = [
            ('CN0 Mean', features.get('cn0_mean', 0), 'dB-Hz'),
            ('CN0 Std', features.get('cn0_std', 0), 'dB-Hz'),
            ('Doppler Rate Std', features.get('doppler_rate_std', 0), 'Hz/s'),
            ('I/Q Correlation', features.get('iq_correlation', 0), ''),
            ('Phase Accel', features.get('phase_acceleration_mean', 0), 'cyc/s²'),
            ('Cycle Slips', features.get('cycle_slip_count', 0), ''),
            ('Spoofing Score', features.get('spoofing_likelihood', 0), '')
        ]
        
        for name, value, unit in key_metrics:
            if unit:
                text += f"{name:18s}: {value:8.3f} {unit}\n"
            else:
                text += f"{name:18s}: {value:8.3f}\n"
        
        return text
    
    def detect_anomalies_sliding_window(self, starting_index=0, window_size=50, step_size=None):
        """
        Sliding window anomaly detection leveraging all available observables.
        
        Parameters:
        -----------
        starting_index : int
            Starting index for analysis
        window_size : int
            Number of samples per window
        step_size : int
            Step size for sliding window (default: window_size//4)
        
        Returns:
        --------
        DataFrame with anomaly scores and individual indicators over time
        """
        if step_size is None:
            step_size = max(1, window_size // 4)
        
        signal_df = self.df.sort_values('time')
        
        if len(signal_df) < window_size:
            print(f"Insufficient data for anomaly detection (need {window_size} samples)")
            return None
        
        results = []
        
        for i in range(starting_index, len(signal_df) - window_size, step_size):
            window = signal_df.iloc[i:i+window_size]
            features = self.compute_enhanced_features(window)
            
            results.append({
                'time': window['time'].iloc[len(window)//2],
                'anomaly_score': features.get('spoofing_likelihood', 0.5),
                'cn0_stability': features.get('cn0_stability', 0),
                'iq_correlation': features.get('iq_correlation', 0),
                'doppler_linearity': features.get('doppler_linearity', 0),
                'pr_doppler_consistency': features.get('pr_doppler_consistency', 0)
            })
        
        results_df = pd.DataFrame(results)
        
        # Visualization
        fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)
        fig.suptitle(f'Continuous Anomaly Detection - Indices {starting_index}+', fontsize=14, fontweight='bold')
        
        # Anomaly score
        ax1 = axes[0]
        ax1.plot(results_df['time'], results_df['anomaly_score'], 'b-', linewidth=2)
        ax1.fill_between(results_df['time'], results_df['anomaly_score'], 0.4, 
                         where=results_df['anomaly_score'] > 0.4,
                         color='orange', alpha=0.3, label='Warning Zone')
        ax1.fill_between(results_df['time'], results_df['anomaly_score'], 0.7,
                         where=results_df['anomaly_score'] > 0.7,
                         color='red', alpha=0.3, label='Critical Zone')
        ax1.axhline(y=0.4, color='orange', linestyle='--', alpha=0.7)
        ax1.axhline(y=0.7, color='red', linestyle='--', alpha=0.7)
        ax1.set_ylabel('Anomaly Score')
        ax1.set_ylim(0, 1)
        ax1.legend(loc='upper right')
        ax1.grid(True, alpha=0.3)
        
        # Individual indicators
        ax2 = axes[1]
        ax2.plot(results_df['time'], results_df['cn0_stability']/20, 
                label='CN0 Stability', alpha=0.8)
        ax2.plot(results_df['time'], np.abs(results_df['iq_correlation']), 
                label='I/Q Correlation', alpha=0.8)
        ax2.plot(results_df['time'], results_df['pr_doppler_consistency'], 
                label='PR-Doppler Consistency', alpha=0.8)
        ax2.set_ylabel('Normalized Indicators')
        ax2.legend(loc='upper right')
        ax2.grid(True, alpha=0.3)
        ax2.set_ylim(0, 1)
        
        # Doppler linearity
        ax3 = axes[2]
        ax3.plot(results_df['time'], results_df['doppler_linearity'], 
                'g-', linewidth=2, label='Doppler Linearity')
        ax3.axhline(y=0.95, color='orange', linestyle='--', 
                   alpha=0.7, label='High Linearity Threshold')
        ax3.set_xlabel('Time (s)')
        ax3.set_ylabel('Linearity (R²)')
        ax3.legend(loc='upper right')
        ax3.grid(True, alpha=0.3)
        ax3.set_ylim(0, 1)
        
        plt.tight_layout()
        plt.show()
        
        return results_df


# Example usage:
if __name__ == "__main__":
    # Initialize analyzer with your dataframe
    analyzer = GNSSSignalAnalyzer(df)
    
    # Analyze authentic signal with I/Q data (using indices)
    print("Analyzing AUTHENTIC signal...")
    result_auth = analyzer.graph_enhanced_stft(
        starting_index=5000,
        ending_index=5100,
        spoofed=None,
        signal_type='iq'
    )
    
    # Analyze spoofed signal (using indices)
    print("\nAnalyzing SPOOFED signal...")
    result_spoof = analyzer.graph_enhanced_stft(
        starting_index=51570,
        ending_index=51670,
        spoofed=None,
        signal_type='iq'
    )

    # Analyze spoofed signal (using indices)
    print("\nAnalyzing SPOOFED signal...")
    result_spoof = analyzer.graph_enhanced_stft(
        starting_index=51670,
        ending_index=51770,
        spoofed=None,
        signal_type='iq'
    )
    
    # Continuous monitoring from a starting index
    print("\nRunning continuous anomaly detection...")
    anomalies = analyzer.detect_anomalies_sliding_window(
        starting_index=0, 
        window_size=50
    )
```
